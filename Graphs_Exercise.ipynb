# Graph 1

```python
import pandas as pd
df = pd.read_csv('C:/Users/mansi/Downloads/AH_Excess_Deaths.csv')
df
```

```python
import matplotlib.pyplot as plt

filtered_data = df[df['AgeGroup'].str.contains('15-19 Years')]
filtered_data['Deaths_Excluding_COVID'] = filtered_data['Deaths'] - filtered_data['Covid19']  # Subtracting COVID-19 deaths

grouped_data = filtered_data.groupby('MMWRyear')['Deaths_Excluding_COVID'].mean() ## Grouping by year and calculating average deaths


plt.figure(figsize=(8, 6))
plt.bar(grouped_data.index, grouped_data.values)

plt.xlabel('Year')
plt.ylabel('Deaths (15-19 Age Group)')
plt.title('Non Covid Deaths in US: 15-19 Years Age Group')

plt.xticks(grouped_data.index, rotation=45, ha='right')
plt.show()
```

# Graph 2

```python
df = pd.read_csv('C:/Users/mansi/Downloads/covid-data.csv')
df
```

```python
df["new_deaths"] = df["total_deaths"].diff() ## Calculating daily new deaths

df["new_deaths_per_million"] = df["new_deaths"] / df["population"] * 1000000  ## Calculating deaths per million
df["new_deaths_per_million"] = df["new_deaths_per_million"].fillna(0)
df['new_deaths_per_million']
```

```python
import matplotlib.dates as mdates

filtered_df = df[(df["date"] >= "2020-01-01") & (df["date"] <= "2021-12-31")]
sweden_data = filtered_df[filtered_df["location"] == "Sweden"][["date", "new_deaths_per_million"]]
uk_data = filtered_df[filtered_df["location"] == "United Kingdom"][["date", "new_deaths_per_million"]]

sweden_data["date"] = pd.to_datetime(sweden_data["date"])
uk_data["date"] = pd.to_datetime(uk_data["date"])

plt.figure(figsize=(8, 6))

# Ploting the Sweden data line
plt.plot(sweden_data["date"], sweden_data["new_deaths_per_million"], label='Sweden')

# Ploting the UK data line
plt.plot(uk_data["date"], uk_data["new_deaths_per_million"], label='UK')

plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('COVID-19 Deaths in Sweden and UK (2020-2021)')

plt.legend()
plt.xticks(rotation=45)
plt.show()
```
